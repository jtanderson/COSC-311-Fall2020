```python
# The python "is" operator: checks if objects are the same (by reference)
x = 10
if x is 10:
    print("Yay!")
```

```python
# Python, for small ints, allocates them as global objects. But for larger objects, they are separate instances
x = 10
y = 10
# while they point to the same object
while x is y:
    print(f"Still the same for {x}")
    x += 1
    y += 1
print(f"Different references fo x: {x}, y: {y}")
```

```python
x = 257
y = 257
if x is y:
    print("Same reference!")
if x == y:
    print("same value!")
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use('seaborn-pastel')


fig = plt.figure()
ax = plt.axes(xlim=(0, 4), ylim=(-2, 2))
line, = ax.plot([], [], lw=3)

def init():
    line.set_data([], [])
    return line,
def animate(i):
    x = np.linspace(0, 4, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return line,

anim = FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=20, blit=True)


anim.save('sine_wave.gif', writer='imagemagick')
```
